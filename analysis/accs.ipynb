```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_acc_dict = np.load(
    '/Users/dg/Projects/task_vectors/task_accs.npy',
    allow_pickle=True
).item()
```

```python
last_layer_test_acc_dict = np.load(
    '/Users/dg/Projects/task_vectors/only_final_class_layer_task_accs.npy',
    allow_pickle=True
).item()
```

```python
cleaned_keys = [float(key) for key in test_acc_dict.keys()]
```

```python
plt.plot(
    cleaned_keys,
    np.array(list(test_acc_dict.values()))[:,0], color='darkblue', label='Accuracy on Task 1'
)
plt.plot(
    cleaned_keys,
    np.array(list(test_acc_dict.values()))[:,1], color='orange', label='Accuracy on Task 2'
)
plt.plot(
    cleaned_keys,
    np.array(list(test_acc_dict.values()))[:,2], label='Multitask accuracy'
)
plt.plot(
    cleaned_keys,
    np.array(list(last_layer_test_acc_dict.values()))[:,0], color='darkblue', label='LL Accuracy on Task 1'
)
plt.plot(
    cleaned_keys,
    np.array(list(last_layer_test_acc_dict.values()))[:,1], color='orange', label='LL Accuracy on Task 2'
)
plt.plot(
    cleaned_keys,
    np.array(list(last_layer_test_acc_dict.values()))[:,2], label='LL Multitask accuracy'
)

plt.hlines(
    max(np.array(list(test_acc_dict.values()))[:,0]), 0,1.0, color='darkblue', linestyle='--', label='Max. task 1'
)
plt.hlines(
    max(np.array(list(test_acc_dict.values()))[:,1]), 0,1.0, color='orange', linestyle='--', label='Max. task 2'
)
plt.xlabel(
    'Model 1 weight'
)
# plt.ylim(
#     0.96,0.99
# )
plt.title(
    'Task 1, 2 and Multitask accuracy'
)
plt.legend(
    loc='upper left'
)
```
