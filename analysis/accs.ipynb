```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_acc_dict = np.load(
    '/Users/dg/Projects/task_vectors/task_accs.npy',
    allow_pickle=True
).item()
```

```python
last_layer_test_acc_dict = np.load(
    '/Users/dg/Projects/task_vectors/only_final_class_layer_task_accs.npy',
    allow_pickle=True
).item()
except_last_layer_test_acc_dict = np.load(
    '/Users/dg/Projects/task_vectors/axcept_last_class_layer_task_accs.npy', 
    allow_pickle=True
).item()
```

```python
cleaned_keys = [float(key) for key in test_acc_dict.keys()]
```

```python
plt.plot(
    cleaned_keys,
    np.array(list(test_acc_dict.values()))[:,0], color='darkblue', label='Accuracy on Task 1'
)
plt.plot(
    cleaned_keys,
    np.array(list(test_acc_dict.values()))[:,1], color='orange', label='Accuracy on Task 2'
)
plt.plot(
    cleaned_keys,
    np.array(list(test_acc_dict.values()))[:,2], color='red', label='Multitask accuracy'
)
plt.plot(
    cleaned_keys,
    np.array(list(last_layer_test_acc_dict.values()))[:,0], color='darkblue', linestyle='--', label='LL Accuracy on Task 1'
)
plt.plot(
    cleaned_keys,
    np.array(list(last_layer_test_acc_dict.values()))[:,1], color='orange', linestyle='--', label='LL Accuracy on Task 2'
)
plt.plot(
    cleaned_keys,
    np.array(list(last_layer_test_acc_dict.values()))[:,2], linestyle='--', color='red', label='LL Multitask accuracy'
)
plt.plot(
    cleaned_keys,
    np.array(list(except_last_layer_test_acc_dict.values()))[:,0], color='darkblue', linestyle='-.', label='!LL Accuracy on Task 1'
)
plt.plot(
    cleaned_keys,
    np.array(list(except_last_layer_test_acc_dict.values()))[:,1], color='orange', linestyle='-.', label='!LL Accuracy on Task 2'
)
plt.plot(
    cleaned_keys,
    np.array(list(except_last_layer_test_acc_dict.values()))[:,2], linestyle='-.', color='red', label='!LL Multitask accuracy'
)
plt.xlabel(
    'Model 1 weight'
)
plt.ylim(
    0.95,1
)
plt.title(
    'Task 1, 2 and Multitask accuracy'
)
plt.legend(
    loc='lower right'
)
```
