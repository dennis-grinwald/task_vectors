```python
import numpy as np
import matplotlib.pyplot as plt
```

# SVHN - Resic

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'brown', 'gray', 'cyan']
for i, j in enumerate([ [ 22 ] , [ 23 ] , [21,22,23] , list(range(23)) ] ):
    if j == list(range(23)):
        path = '../results/accs_simplex_in_layers_all.npy'
    else: 
        path = f'../results/accs_simplex_in_layers_{"_".join(map(str, j))}.npy'
    test_acc_dict = np.load(
        path,
        allow_pickle=True
    ).item()
    cleaned_keys = [float(key) for key in test_acc_dict.keys()]
    ax[0].plot(
        cleaned_keys,
        np.array(list(test_acc_dict.values()))[:,0], linestyle='--', color=colors[i], label=f'Layer{j}'
    )
    ax[1].plot(
        cleaned_keys,
        np.array(list(test_acc_dict.values()))[:,1], linestyle='-.', color=colors[i], label=f'Layer{j}'
    )
ax[0].set_title('Task 1')
ax[1].set_title('Task 2')
ax[0].legend(
    loc='lower right'
)
ax[1].legend(
    loc='lower left'
)
```
